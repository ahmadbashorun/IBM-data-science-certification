## Exploratory Data Analysis


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
from js import fetch
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(dataset_part_2_csv)
df.head(5)
```

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass also appears to be a factor; even with more massive payloads, the first stage often returns successfully.


```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

Next, let's drill down to each site visualize its detailed launch records.


### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
print("\n" + "="*50)
print("TASK 1: Flight Number vs Launch Site")
print("="*50)

plt.figure(figsize=(12, 6))
sns.catplot(x="FlightNumber", y="LaunchSite", hue="Class", data=df, aspect=3)
plt.xlabel("Flight Number", fontsize=16)
plt.ylabel("Launch Site", fontsize=16)
plt.title("Relationship between Flight Number and Launch Site", fontsize=18)
plt.tight_layout()
plt.show()

print("\nPattern Analysis:")
print("- CCAFS SLC 40 was used for earlier flights (Flight 1-60)")
print("- KSC LC 39A started being used around Flight 60")
print("- VAFB SLC 4E appears intermittently throughout")
print("- Success rate improves over time across all launch sites")
```

Now try to explain the patterns you found in the Flight Number vs. Launch Site scatter point plots.


### TASK 2: Visualize the relationship between Payload Mass and Launch Site


We also want to observe if there is any relationship between launch sites and their payload mass.


```python
print("\n" + "="*50)
print("TASK 2: Payload Mass vs Launch Site")
print("="*50)

plt.figure(figsize=(12, 6))
sns.catplot(x="PayloadMass", y="LaunchSite", hue="Class", data=df, aspect=3)
plt.xlabel("Payload Mass (kg)", fontsize=16)
plt.ylabel("Launch Site", fontsize=16)
plt.title("Relationship between Payload Mass and Launch Site", fontsize=18)
plt.tight_layout()
plt.show()

print("\nPattern Analysis:")
print("- VAFB SLC 4E: No launches with payload > 10,000 kg")
print("- KSC LC 39A: Handles the heaviest payloads (up to 15,000+ kg)")
print("- CCAFS SLC 40: Medium range payloads")
```

Now if you observe Payload Mass Vs. Launch Site scatter point chart you will find for the VAFB-SLC  launchsite there are no  rockets  launched for  heavypayload mass(greater than 10000).


### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
print("\n" + "="*50)
print("TASK 3: Success Rate by Orbit Type")
print("="*50)

# Calculate success rate by orbit
orbit_success = df.groupby('Orbit')['Class'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
orbit_success.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel("Orbit Type", fontsize=14)
plt.ylabel("Success Rate", fontsize=14)
plt.title("Success Rate by Orbit Type", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1.1)

# Add value labels on bars
for i, (orbit, rate) in enumerate(orbit_success.items()):
    plt.text(i, rate + 0.02, f'{rate:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

print("\nOrbit Success Rates:")
for orbit, rate in orbit_success.items():
    print(f"- {orbit}: {rate:.2%}")
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.


### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# TASK 4: Visualize the relationship between FlightNumber and Orbit type
print("\n" + "="*50)
print("TASK 4: Flight Number vs Orbit Type")
print("="*50)

plt.figure(figsize=(14, 8))
sns.scatterplot(x="FlightNumber", y="Orbit", hue="Class", data=df, s=100, alpha=0.8)
plt.xlabel("Flight Number", fontsize=14)
plt.ylabel("Orbit Type", fontsize=14)
plt.title("Relationship between Flight Number and Orbit Type", fontsize=16)
plt.legend(title='Landing Success', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()

print("\nPattern Analysis:")
print("- LEO orbits show increasing success rate with flight number")
print("- GTO orbits show mixed results throughout")
print("- ISS missions appear mostly in later flights with high success")
```

You can observe that in the LEO orbit, success seems to be related to the number of flights. Conversely, in the GTO orbit, there appears to be no relationship between flight number and success.


### TASK  5: Visualize the relationship between Payload Mass and Orbit type


Similarly, we can plot the Payload Mass vs. Orbit scatter point charts to reveal the relationship between Payload Mass and Orbit type


```python
# TASK 5: Visualize the relationship between Payload Mass and Orbit type
print("\n" + "="*50)
print("TASK 5: Payload Mass vs Orbit Type")
print("="*50)

plt.figure(figsize=(14, 8))
sns.scatterplot(x="PayloadMass", y="Orbit", hue="Class", data=df, s=100, alpha=0.8)
plt.xlabel("Payload Mass (kg)", fontsize=14)
plt.ylabel("Orbit Type", fontsize=14)
plt.title("Relationship between Payload Mass and Orbit Type", fontsize=16)
plt.legend(title='Landing Success', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()

print("\nPattern Analysis:")
print("- Polar, LEO, and ISS orbits show better success with heavier payloads")
print("- GTO shows mixed results across all payload masses")
print("- Different orbits have different typical payload mass ranges")
```

With heavy payloads the successful landing or positive landing rate are more for Polar,LEO and ISS.

However, for GTO, it's difficult to distinguish between successful and unsuccessful landings as both outcomes are present.


### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
print("\n" + "="*50)
print("TASK 6: Launch Success Yearly Trend")
print("="*50)

# Extract year from date
year = []
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year

Extract_year()
df['Date'] = year
df.head()

# Calculate yearly success rate
yearly_success = df.groupby('Date')['Class'].mean()

plt.figure(figsize=(12, 6))
yearly_success.plot(kind='line', marker='o', markersize=8, linewidth=2, color='darkblue')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Success Rate", fontsize=14)
plt.title("Launch Success Rate Trend Over Years", fontsize=16)
plt.grid(True, alpha=0.3)
plt.ylim(0, 1.1)

# Add value labels
for year, rate in yearly_success.items():
    plt.text(year, rate + 0.02, f'{rate:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

print("\nYearly Success Rates:")
for year, rate in yearly_success.items():
    print(f"- {year}: {rate:.2%}")
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
```

you can observe that the sucess rate since 2013 kept increasing till 2020


## Features Engineering


By now, you should obtain some preliminary insights about how each important variable would affect the success rate, we will select the features that will be used in success prediction in the future module.


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# One-hot encode categorical columns
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

print("\nShape before encoding:", features.shape)
print("Shape after encoding:", features_one_hot.shape)
print("\nFirst 5 rows of encoded dataframe:")
print(features_one_hot.head())

print("\nNew columns created:")
new_cols = [col for col in features_one_hot.columns if col not in features.columns]
print(f"Total new columns: {len(new_cols)}")
for i, col in enumerate(new_cols[:10]):  # Show first 10
    print(f"  - {col}")
if len(new_cols) > 10:
    print(f"  ... and {len(new_cols) - 10} more columns")
```

### TASK  8: Cast all numeric columns to `float64`


Now that our <code>features_one_hot</code> dataframe only contains numbers, cast the entire dataframe to variable type <code>float64</code>


```python
# Convert all columns to float64
features_one_hot = features_one_hot.astype('float64')

print("\nData types after conversion:")
print(features_one_hot.dtypes.value_counts())
print("\nAll columns are now type:", features_one_hot.dtypes.unique()[0])

# Save the prepared features
features_one_hot.to_csv('dataset_part_3.csv', index=False)
print("\n✅ Feature engineering complete! Data saved to 'dataset_part_3.csv'")
print(f"Final dataset shape: {features_one_hot.shape}")
```

```python
# Summary statistics
print("\n" + "="*50)
print("SUMMARY")
print("="*50)
print(f"Total launches analyzed: {len(df)}")
print(f"Overall success rate: {df['Class'].mean():.2%}")
print(f"Total features after encoding: {features_one_hot.shape[1]}")
print("\nKey Insights:")
print("1. Success rate has improved significantly over time (from ~0% to ~80%)")
print("2. Different launch sites are optimized for different payload masses")
print("3. LEO, ISS, and Polar orbits have higher success rates than GTO")
print("4. Later flights (higher flight numbers) generally have better success rates")
print("5. VAFB SLC 4E is limited to lighter payloads (<10,000 kg)")
```
